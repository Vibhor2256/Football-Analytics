```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load dataset
epl_df=pd.read_csv('/content/2021-22_EPL_PlayerStats_updated new - 2021-22_EPL_PlayerStats_updated.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
epl_df.isna().sum()
```

```python
#Create 2 new columns

epl_df['MinsPerMatch']= (epl_df['Mins']/epl_df['Matches'])
epl_df['GoalsPerMatch']= (epl_df['Goals']/epl_df['Matches'])
```

```python
epl_df.head()
```

```python
#Total Goals

ttlGoals=epl_df['Goals'].sum()
ttlGoals
```

```python
#Penanlty Goals

ttlPenalty=epl_df['PnG'].sum()
ttlPenalty
```

```python
#Penalty Attempts

ttlPnAtt=epl_df['PnAtt'].sum()
ttlPnAtt
```

```python
#Pie chart for penalties missed vs scored

plt.figure(figsize=(13,6))
Pn_not_scored=ttlPnAtt - ttlPenalty
data= [Pn_not_scored, ttlPenalty]
labels= ['Penalty Missed', 'Penalty Scored']
color= sns.color_palette('Set2')
plt.pie(data, labels= labels, colors= color, autopct= '%.2f%%')
plt.show()

```

```python
#Most players from which countires

max_natnlty= epl_df.groupby('Nationality').size().sort_values(ascending=False)
max_natnlty.head(10).plot(kind='bar', figsize=(12,6), color= sns.color_palette('magma'))
```

```python
#Clubs with maximum players in their squad

epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', color= sns.color_palette('viridis'))
```

```python
#Clubs with minimum players in their squad

epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color= sns.color_palette('viridis'))
```

```python
#Total Assists from each club

Assists_by_clubs= pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
var=sns.barplot(x= 'Club', y= 'Assists', data= Assists_by_clubs.sort_values(by='Assists'), palette='Set2')
var.set_xlabel("Club", fontsize=30)
var.set_ylabel("Assists", fontsize=25)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']= (16,10)
#plt.figure(figsize=(8,6))
plt.title('Plot of Clubs vs Total Assists', fontsize=20)
```

```python
#Top 10 Assists

top10_assists= epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n=10, columns= 'Assists')
top10_assists
```

```python
Goals_by_clubs=  pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
var2=sns.barplot(x= 'Club', y= 'Goals', data= Goals_by_clubs.sort_values(by='Goals'), palette='rocket')
var2.set_xlabel("Club", fontsize=30)
var2.set_ylabel("Goals", fontsize=25)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']= (16,10)
#plt.figure(figsize=(8,6))
plt.title('Plot of Clubs vs Total Goals', fontsize=20)
```

```python
# Most goals by players

top10_goals= epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(n=10, columns= 'Goals')
top10_goals
```

```python
# Goals per match

top10_goalsPerMatch= epl_df[['Name', 'GoalsPerMatch', 'Matches', 'Goals']].nlargest(n=10, columns= 'GoalsPerMatch')
top10_goalsPerMatch
```

```python
#Pie chart - Goals with assist and without assist

plt.figure(figsize=(14,8))
assists= epl_df['Assists'].sum()
data= [ttlGoals - assists, assists]
labels= ['Goals without assists', 'Goals with assists']
color= sns.color_palette('Set1')
plt.pie(data, labels= labels, colors= color, autopct= '%.0f%%')
plt.show()
```

```python
#Top 10 players with highest yellow card

epl_yellow= epl_df.sort_values(by='YC', ascending= False)[:10]
plt.figure(figsize=(14,8))
plt.title('Players with highest yellow cards')
var3=sns.barplot(x=epl_yellow['Name'], y=epl_yellow['YC'], label='Players', color='Yellow')
plt.ylabel('Number of Yellow Cards')
var3.set_xticklabels(var3.get_xticklabels(), rotation= 45)
var3
```
